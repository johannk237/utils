```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
names=[]
details=[]
total_pages = 60
for page_num in range(1, total_pages + 1):
    url = f'https://onmc.app/tableau_de_lordre?page={page_num}'

    # Envoyer une requête GET à la page
    response = requests.get(url)
    response.raise_for_status()  # Vérifie que la requête a réussi
    
    # Analyser le contenu HTML de la page
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Trouver toutes les entrées de médecins
    # Supposons que chaque médecin est dans une balise <div class="bloc-containt">
    doctor_entries = soup.find_all('div', class_='bloc-containt')
    
    for entry in doctor_entries:
            name = entry.find('h2').get_text(strip=True)
            detail = entry.find('span').get_text(strip=True)
            names.append(name)
            details.append(detail)
            print(f'Nom: {name}, Détails: {detail}')

pd.DataFrame({
    "nom":names,
    "details": details,
}
            ).to_csv("medecins.csv", index=False)
```
